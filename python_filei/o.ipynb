```python
#SAMPLE QUESTION 1:
# import re
# fs = open("text.txt","r")
# fd = open("text2.txt","w")
# lines = fs.readlines()
# header = lines[0].strip()+",Tax\n"
# fd.write(header)
# for i in lines[1:]:
#     parts = i.strip().split(",")
#     #emp ID
#     parts[0] = re.sub(r"[@#]+","-",parts[0])
#     #name
#     if parts[1] == "":
#         parts[1] = "unknown"
#     #salary
#     try:
#         salary = float(parts[3])
#     except ValueError:
#         salary = 0.0
#     salary = round(salary)
#     parts[3]=str(salary)
#     #email
#     parts[4] = parts[4].replace("@@","@")
#     if not parts[4].endswith(".com"):
#         parts[4]+=".com"
#     if(salary>=80000):
#         parts.append("High")
#     elif(50000<=salary<80000):
#         parts.append("Medium")
#     else:
#         parts.append("Low")

#     new_line = ",".join(parts)
#     fd.write(new_line+"\n")
# fs.close()
# fd.close()

# Q2. Student Grade Processing
# ðŸ“‚ File: students.tx
# Columns: ID, Name, Marks, Department
# Tasks:
# Clean ID (remove spaces/symbols).
# If Name empty â†’ "Unknown".
# Convert Marks â†’ integer (if invalid â†’ 0).
# Add a new column Grade:
# A if marks â‰¥ 85
# B if 70â€“84
# C if 50â€“69
# F if < 50
# Save cleaned data to students_cleaned.txt.
import re
with open("student_data_txt.txt","r") as f:
  lines=f.readlines()
  header=lines[0].strip()+",Grade\n"
  for i in lines[1:]:
    parts=i.strip().split(",")
    #clean_id
    parts[0]=re.sub(r'[^a-zA-Z0-9]',"",parts[0].strip())
    #clean_name
    if(parts[1]==""):
      parts[1]="unknown"
    #clean_marks
    try:
      marks=int(parts[2])
    except ValueError:
      marks=0
    parts[2]=str(marks)
    #clean_dept
    parts[3]=parts[3].strip()
    #clean_grade
    if(marks>=85):
      parts.append("A")
    elif(70<=marks<=84):
      parts.append("B")
    elif(50<=marks<=69):
      parts.append("C")
    else:
      parts.append("F")
    new_line=",".join(parts)
with open("student_cleaned.txt","w") as f1:
  f1.write(new_line+"\n")



# Q3. Book Store Sales Normalization

# ðŸ“‚ File: books.txt

# Columns: BookID, Title, Author, Price, Copies

# Tasks:

# Fix BookID (B001 format).

# Remove extra spaces/special chars in Title.

# Ensure Price is float, round to 2 decimals.

# Ensure Copies is integer (default 0).

# Add new column TotalSales = Price * Copies.

# Write to books_cleaned.txt.

# Q4. Hospital Patient Data Cleaning

# ðŸ“‚ File: patients.txt

# Columns: PatientID, Name, Age, Disease, Bill

# Tasks:

# Format PatientID as P-000X.

# Replace missing Name with "Anonymous".

# Ensure Age is integer (default 0).

# Remove digits from Disease field (only letters allowed).

# Add Category: "Senior" if Age â‰¥ 60 else "Adult".

# Save to patients_cleaned.txt.

# Q5. Movie Ratings Transformation

# ðŸ“‚ File: movies.txt

# Columns: MovieID, Title, Genre, Rating, Year

# Tasks:

# Format MovieID â†’ M-XXX.

# Ensure Rating is float (default 0.0).

# If Year < 2000 â†’ add Category "Classic", else "Modern".

# Ensure Genre abbreviations: Actionâ†’ACT, Dramaâ†’DRM, Comedyâ†’COM.

# Save to movies_cleaned.txt.

# Q6. Product Catalog Cleanup

# ðŸ“‚ File: products.txt

# Columns: ProdID, Name, Category, Price, Discount%

# Tasks:

# Format ProdID â†’ PRD-XXX.

# Ensure Price is float (default 0.0).

# Ensure Discount% is integer (default 0).

# Add column FinalPrice = Price - (Price * Discount/100).

# Save to products_cleaned.txt.

# Q7. Library Member Database

# ðŸ“‚ File: members.txt

# Columns: MemberID, Name, Age, Email, BooksIssued

# Tasks:

# Fix MemberID format â†’ MBR-XXX.

# If Email missing â†’ "not_provided@email.com".

# Age must be integer (default 0).

# If BooksIssued > 5 â†’ add "VIP" else "Normal".

# Save to members_cleaned.txt.

# Q8. Transport Bus Records

# ðŸ“‚ File: buses.txt

# Columns: BusID, DriverName, Route, Capacity, Occupied

# Tasks:

# Format BusID â†’ BUS-XXX.

# Ensure Capacity/Occupied are integers (default 0).

# Calculate VacantSeats = Capacity - Occupied.

# If VacantSeats < 0 â†’ make 0.

# Save to buses_cleaned.txt.

# Q9. Online Orders Management

# ðŸ“‚ File: orders.txt

# Columns: OrderID, CustomerName, Item, Price, Quantity

# Tasks:

# Format OrderID â†’ ORD-XXX.

# Clean CustomerName (no digits/special chars).

# Ensure Price is float, Quantity is int.

# Add new column TotalAmount = Price * Quantity.

# Save to orders_cleaned.txt.

# Q10. Exam Results Categorization

# ðŸ“‚ File: results.txt

# Columns: RollNo, Name, Subject, Marks

# Tasks:

# Format RollNo â†’ R-XXXX.

# If Name empty â†’ "Unknown".

# Marks must be int (default 0).

# Add Result = Pass if Marks â‰¥ 40 else Fail.

# Save to results_cleaned.txt.
```
