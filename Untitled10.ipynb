```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import *
spark = SparkSession.builder.appName("CSV Reader").getOrCreate()

df1=spark.read.csv("customers.csv",inferSchema=True,header=True,sep=",")
df2=spark.read.csv("products.csv",inferSchema=True,header=True,sep=",")
df3=spark.read.csv("orders.csv",inferSchema=True,header=True,sep=",")
df4=spark.read.csv("order_items.csv",inferSchema=True,header=True,sep=",")
df5=spark.read.csv("deliveries.csv",inferSchema=True,header=True,sep=",")
df1.show()
df2.show()
df3.show()
df4.show()
df5.show()
df1.createOrReplaceTempView("cust")
df2.createOrReplaceTempView("prod")
df3.createOrReplaceTempView("ord")
df4.createOrReplaceTempView("ord_items")
df5.createOrReplaceTempView("del")

```

```python
j1=df1.join(df3,df1.customer_id==df3.customer_id,"inner")
j2=j1.join(df4,j1.order_id==df4.order_id,"inner")
j3=j2.join(df2,j2.product_id==df2.product_id,"inner")
j3.show()
j1.createOrReplaceTempView("j1")
j2.createOrReplaceTempView("j2")
j3.createOrReplaceTempView("j3")
```

```python
# Find the total revenue generated by each city.
spark.sql("select city,sum(price) as total_revenue from j3 group by city").show()
# Get the top 5 customers who spent the most across all orders.
spark.sql("select c.name as customer_name,sum(p.price*oi.quantity) as total_spent from cust c join ord o on o.customer_id=c.customer_id join ord_items oi on oi.order_id=o.order_id join prod p on p.product_id=oi.product_id group by customer_name order by total_spent desc limit 5").show()
# Find the average order value per customer.
spark.sql("select c.name as customer_name,avg(p.price*oi.quantity) as avg_order_value from cust c join ord o on o.customer_id=c.customer_id join ord_items oi on oi.order_id=o.order_id join prod p on p.product_id=oi.product_id group by customer_name").show()
# Count the number of distinct customers per state.
spark.sql("select count(distinct c.customer_id) as unique_customers,state from cust c group by state ").show()
# Find the most popular product category in each city.
spark.sql("select c.city,p.category,sum(oi.quantity) as popular_product from cust c join ord o on o.customer_id=c.customer_id join ord_items oi on oi.order_id=o.order_id join prod p on p.product_id=oi.product_id group by c.city,p.category order by c.city,popular_product desc").show()
# List customers who have placed more than 5 orders.
spark.sql("select c.name as customer_name,sum(oi.quantity) as total_orders from cust c join ord o on o.customer_id=c.customer_id join ord_items oi on oi.order_id=o.order_id group by customer_name having total_orders>5").show()

# joined_final=spark.sql("select * from cust c join ord o on o.customer_id=c.customer_id join ord_items oi on oi.order_id=o.order_id").show()
# joined_final.createOrReplaceTempView("joined_final")
# Find orders with more than 3 unique products.
spark.sql("select o.order_id,count(distinct oi.product_id) as unique_products from ord o join ord_items oi on oi.order_id=o.order_id group by o.order_id having count(distinct oi.product_id)>3").show()
# Find customers who never placed an order.
spark.sql("select c.name as customer_name,o.order_id from cust c left join ord o on c.customer_id=o.customer_id where o.order_id is null").show()
# Calculate the average delivery time (in days) per courier.
spark.sql("select d.courier, avg(datediff(d.delivery_date,o.order_date)) as avg_delivery_time from del d join ord o on o.order_id=d.order_id where d.delivery_date is not null group by d.courier").show()
# Find the city with the highest number of pending deliveries.
spark.sql("select c.city,count(*) as pending_deliveries from cust c join ord o on o.customer_id=c.customer_id join del d on d.order_id=o.order_id where d.delivery_status='Pending' group by c.city order by pending_deliveries desc limit 1").show()
# Identify customers with total purchases > ₹1,00,000.
spark.sql("select c.name,sum(p.price*oi.quantity) as total_purchase from cust c join ord o on c.customer_id=o.customer_id join ord_items oi on oi.order_id=o.order_id join prod p on p.product_id=oi.product_id group by c.name having total_purchase >100000 order by total_purchase desc").show()
# Find orders where at least one item price > 50,000.
spark.sql("select o.order_id,p.price from ord o join ord_items oi on oi.order_id=o.order_id join prod p on p.product_id=oi.product_id where p.price>50000").show()
# Get the most popular product in each category.
spark.sql("select p.product_id,p.category,sum(oi.quantity) as popular_product from prod p join ord_items oi on oi.product_id=p.product_id group by p.product_id,p.category order by popular_product desc").show()
# Count orders by payment method.
spark.sql("select payment_method,count(*) from ord group by payment_method").show()
# Get all orders placed within the last 30 days.
spark.sql("select o.order_id,o.order_date,c.name as customer_name from cust c join ord o on o.customer_id=c.customer_id where datediff(current_date(),order_date)<=30").show()
# Find products that have never been ordered.
spark.sql("select p.product_id,o.order_id from prod p left join ord_items oi on oi.product_id=p.product_id left join ord o on o.order_id=oi.order_id where o.order_id is null").show()
```

```python
# Here are 4 Spark SQL practice questions on CONCAT (only questions, no solutions):

# Create a full customer address by concatenating city, state, and country into a single column.
spark.sql("select name,concat(city,',',state,',',country) as full_address from cust").show()
# Generate unique order reference IDs by concatenating order_id with customer_id.
spark.sql("select order_id,customer_id,concat(order_id,'-',customer_id) as unique_id from ord").show()
# Show customer name along with their email domain in the format "Name (domain)".
spark.sql("select name,concat(name,'(',split(email,'@')[1],')') as name_email from cust").show()
# Create a delivery summary message like: “Order ORD1001 for Raj Shekhar Singh is Delivered by
spark.sql("select concat('Order ',o.order_id,' for',c.name,' is Delivered by ',d.courier) as delivery_message from cust c join ord o on o.customer_id=c.customer_id join del d on d.order_id=o.order_id").show()
```
